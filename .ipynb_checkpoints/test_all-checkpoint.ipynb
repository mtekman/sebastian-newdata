```r
sce <- calculateQCMetrics(sce)
hist(sce$total_counts, breaks = 100)
ab_gruen = 3000
abline(v = ab_gruen, col = "blue")

num_cells_above_depth <- function(object, depth){
    return(dim(object[,colSums(counts(object)) > depth])[[2]])
}

determine_maxseqdepth_beforeloss <- function(obj)
{
    # At bare minimum we want over 0 total trans.
    original_num <- num_cells_above_depth(obj, 0)     
    
    running_depth = 0
    iterate_step = 5000
       
    while (iterate_step > 100){
        current_count  = num_cells_above_depth(obj, running_depth)
        
        #message(current_count, "     ", running_depth, "     ", iterate_step)
              
        # Reduction, we need to step back 1/2
        if (original_num > current_count){
            running_depth = running_depth - iterate_step
            iterate_step = as.integer(iterate_step / 2)
        }
        
        running_depth = running_depth + iterate_step       
    }
    return(running_depth)
}

cutoff <- determine_maxseqdepth_beforeloss(sce)
abline(v = cutoff, col = 'red')

message("Max seq depth before drop in cell numbers: ", cutoff)


#hist(sce$total_features, breaks = 100)
#abline(v = 3500, col = "red")
#cells_filtered_hem <- 
#cells_filtered_hem
```

```r
# As we can see, RaceID3 with reduced minexpr settings perfectly encapsulates RaceID2 with defaults
# (and also RaceID3 with defaults)
#
# Grün's method of normalisation (doi:10.1038/nature15966):
# - Minimum of 3000 transcripts per cells
# - Then median-based( only raceID2, raceID3 uses rescaling by minimum) normalisation across cells OR downsampling to min transcript
# - A pseudocount of 0.1 is added to all
# - Gene filtering then takes place:
#    - "Lowly expressed genes are filtered out": Genes not expressed across a minimum of 5 transcripts in the whole organoids data are removed in at least 1 cell
#    - Highly expressed genes with more than 500 transcripts are normalisation are also removed

# Let us now recapitulate these results using default SCATER filtering
# --why ? In order to work with the SingleCellExperiment object of course!

dim(sce)
min_3k <- sce[,sce$total_counts > 3000]
dim(min_3k)


# Normalization is performed by dividing transcript counts in each cell by the total number of transcripts in 
# this cell followed by multiplication with the minimum of the total number of transcripts across cells 
# (normalization by rescaling)

raceID3_rescaling_norm <- function(obj){
    count_data <- t(t(obj) / colSums(obj))      # Divide all transcripts by their column totals (the double transpose is weirld neccesary)
    min_transcript_total <- min(colSums(obj))   # Minumum transcript total across all cells
    
    count_data <- count_data * min_transcript_total
    return(count_data + 0.1)
}

raceID3_gene_filtering <- function(dat, minexp, maxexp, minnumber){
    #dat <- dat[rowSums(dat) > minexp,]   # Filter minimum expression
    #message("Remaining after filtering for expression > ", minexp, " = ", dim(dat)[[2]])
    dat <- dat[rowSums(dat) <= maxexp,]   # Filter maximum expression
    message("Remaining after filtering for expression <= ", maxexp, " = ", dim(dat)[[2]])
    gene_detectability <- apply(dat, 1, function(row){
        num_non_zero_cells = 0
        for (element in row){
            if (element >= minexp){
                num_non_zero_cells = num_non_zero_cells + 1
            }
        }
        return(num_non_zero_cells)
    })
    #return(gene_detectability > 0)
    return(dat[gene_detectability >= minnumber,])
    
}

head(r3obj_f_extra@fdata)
rescaled <- raceID3_rescaling_norm(counts(sce))
head(rescaled)
genefiltered <- raceID3_gene_filtering(rescaled, 1, 500, 1)
head(genefiltered)

# The normalisation now matches, but the gene filtering is off
# Discrepancy between r3obj_f_extra and rescaled[genefiltered,] ... !

## This workflow is not important. Why am I trying to replicated RaceID3 in SCATER? Why not just use RaceID3?
## We want to analyse the data, we don't care about anything else!





```
