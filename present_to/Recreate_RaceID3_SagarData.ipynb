```r
out <- rbind(
    dim(sc@fdata),
    dim(sc3@fdata),
    dim(sc3Inf@fdata),
    dim(sc4@fdata),
    dim(sc4Inf@fdata),
    dim(sc5@fdata),
    dim(sc5Inf@fdata),
    dim(control_sc@fdata)
)
rownames(out) <- c("sc","sc3","sc3+Inf", "sc4", "sc4+Inf", "sc5", "sc5+Inf", "Sagar")
colnames(out) <- c("Genes", "Cells")

out
```

### ccCleaner + Biomart

```r
require(biomaRt)
# Annotate the data before adding SC experiment

mart <- useMart(biomart = "ENSEMBL_MART_ENSEMBL", 
                dataset = "mmusculus_gene_ensembl")

k <- getBM(
    filters = "mgi_symbol",
    attributes = c(
            "ensembl_gene_id",              # Gene stable ID
            "external_gene_name",           # Casual name
            "mgi_symbol",
#            "external_transcript_name",     # Transcript-specific name
#            "gene_biotype",                 # Gene biotype
#            "transcript_biotype",           # Trans type
#            "description",                  # Gene description
#            "band",                         # Karyotype band
#            "refseq_mrna",
            "go_id",                        # Go Term accession (cellular domains)
#            "go_linkage_type",              # Go Term evidence code
            "name_1006"                    # Go Term name
#            "definition_1006",              # Go Term definition
#            "namespace_1003"                # Go domain                  
    ),
    values = rownames(sc5@fdata),
    mart = mart
)
```

```r
gCC <- k$mgi_symbol[k$name_1006 == "cell cycle"]
gCP <- k$mgi_symbol[k$name_1006 == "cell proliferation"]
length(gCC)
length(gCP)

sagar_cGenes = c("Malat1", "Lars2", "Mki67", "Pcna-ps2")
sagar_fGenes = c("Hsp90aa1", "Jun", "Fos")

sagar_cGenes %in% gCC
sagar_cGenes %in% gCP
sagar_fGenes %in% gCC
sagar_fGenes %in% gCP

# Mki67 -- ENSG00000148773
# head(tail(k, -10700),50) #-- nothing
# grep("ENSG00000148773", k) #-- nothing

# Hsp90aa1 -- ENSMUSG00000021270
hsp_subk <- k[k$mgi_symbol %in% "Hsp90aa1",]
hsp_subk[grepl("prolif", hsp_subk$name_1006),] #--nothing
# This gene is in our matrix, but it is NOT flagged for cell proliferation,
# only the human transcript for this is, but not the mouse.
# (human:ENSG00000080824 → evidence:GO:0008284, 
#  mouse:ENSMUSG00000021270 → evidence:none )
```

### Perform K-medoids clustering

```r
# Sagar's settings
rbind(control_sc@clusterpar)
```

He uses all the defaults except cln (number of k clusters) which is normally 0 and is derived from using the do.gap = T. He hasn't set do.gap, so it's strange to see how he came up with 12.

Also the SE.method is not "Tibs2001SEmax" but "firstSEmax"

Let's try running with defaults on our own data and see if we can recapitulate his settings.

### Defaults with gap

2 clusters are suggested instead of 12. Odd...

```r
zzz <- capture.output(sc_e <- clustexp(sc_e, do.gap = T))

rbind(sc_e@clusterpar)
par(mfrow=c(2,2))
plotsaturation(sc_e,disp=F)
plotgap(sc_e)
plotjaccard(sc_e)
plotsilhouette(sc_e)

x <- clustheatmap(sc_e, final=F, hmethod="single")
```

### With firstSEMax and gap

2 Clusters are again suggested, instead of 12. Weak.

Change in method showed no difference, but we can see from the plateau that the Gap is around 12.

```r
zzz <- capture.output(sc_e <- clustexp(sc_e, do.gap = T, SE.method = "firstSEmax"))

rbind(sc_e@clusterpar)
par(mfrow=c(2,2))
plotsaturation(sc_e,disp=F)
plotgap(sc_e)
plotjaccard(sc_e)
plotsilhouette(sc_e)

x <- clustheatmap(sc_e, final=F, hmethod="single")
```

This method took twice as long and does not appear to improve the expected cluster number of 12.

### Proceeding with 12

I can now understand why the defaults were not used -- namely that the number of predicted clusters was far too low.

```r
zzz <- capture.output(sc_e <- clustexp(sc_e, do.gap = F, SE.method = "firstSEmax", cln = 12))
```

```r
rbind(sc_e@clusterpar)
par(mfrow=c(1,2))
plotsaturation(sc_e,disp=F)
plotjaccard(sc_e)
#plotgap(sc_e)

plotsilhouette(sc_e)
x <- clustheatmap(sc_e, final=F, hmethod="single")
```
